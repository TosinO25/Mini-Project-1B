## Part 5: Extract Features

Create a set of features that capture important information about the EEG signals. Common features include:
- Power in each frequency band
- Statistical measures (mean, variance, skewness)
- Entropy measures
- Band power ratios

Build a feature matrix where each row is a sample and each column is a feature.

```python
def extract_features(eeg_signal, sampling_rate=256):
    """Extract features from an EEG signal.
    
    Parameters:
    - eeg_signal: 2D array (samples, channels)
    - sampling_rate: frequency in Hz
    
    Returns:
    - features: 1D array of extracted features
    """
    
    features = []
    
    # TODO: Extract features from the signal
    # Example features to try:
    # 1. Power in alpha band (8-12 Hz)
    # 2. Power in beta band (12-30 Hz)
    # 3. Power in gamma band (30-50 Hz)
    # 4. Mean absolute value
    # 5. Standard deviation
    # 6. Variance
    # 7. Entropy
    
    # Start simple, then add more features if needed
    for channel in range(eeg_signal.shape[1]):
        signal_channel = eeg_signal[:, channel]
        
        # Statistical features
        features.append(np.mean(signal_channel))
        features.append(np.std(signal_channel))
        features.append(np.var(signal_channel))
        
        # Frequency domain features
        freq, psd = signal.welch(signal_channel, fs=sampling_rate, nperseg=256)
        
        # Power in frequency bands
        alpha_power = np.sum(psd[(freq > 8) & (freq < 12)])
        beta_power = np.sum(psd[(freq > 12) & (freq < 30)])
        gamma_power = np.sum(psd[(freq > 30) & (freq < 50)])
        
        features.extend([alpha_power, beta_power, gamma_power])
    
    return np.array(features)

# TODO: Apply to all samples
# feature_matrix = np.array([extract_features(sample) for sample in your_data])

print('Feature extraction function defined.')
print('Next: apply this to all samples in your dataset')
```
